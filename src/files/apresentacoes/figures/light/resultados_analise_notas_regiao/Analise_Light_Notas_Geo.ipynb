## Análises de Litigância por tipos de notas - LIGHT

```python
import pandas as pd
import numpy as np
import pylab
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
```

```python
datapath = "/home/rsouza/Dropbox/Renato/Light/"
```

```python
df = pd.read_csv(datapath+'all.csv', 
                 names=['processo','toi','negativacao','corte','reclamacao','quantidade'], 
                 skiprows = 1)
df[:10]
```

```python
df.describe()
```

```python
df_processo = df.sort(['processo'], ascending=1)
regiao = df_processo.index
percent_processos = df_processo.processo * 100
y_pos = np.arange(len(df_processo))

plt.figure(figsize=(10, 10))
plt.barh(y_pos, percent_processos)
plt.yticks(y_pos+0.5, regiao, fontsize=9, family='sans-serif')
plt.xlabel('Percentual do total')
plt.title(u'Percentual de Processos por Instalação nas Regiões')

#add the numbers to the side of each bar
for p, ch in zip(y_pos, percent_processos):
    plt.annotate(str(round(ch, ndigits=2)), xy=(ch + 0.1, p + .5), va='center')

#set plot limits
plt.ylim(y_pos.max(), y_pos.min() - 1)
plt.xlim(0, 1.8)

plt.show()
```
